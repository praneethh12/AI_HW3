```python
import os
import subprocess

input_dir = "videos"     # folder containing all .mp4 files
frames_dir = "frames2"    # where all extracted frames will go

os.makedirs(frames_dir, exist_ok=True)

for fname in os.listdir(input_dir):
    if not fname.endswith(".mp4"):
        continue

    video_path = os.path.join(input_dir, fname)
    base = os.path.splitext(fname)[0]

    # Make a subfolder for this video’s frames (optional but cleaner)
    out_dir = os.path.join(frames_dir, base)
    os.makedirs(out_dir, exist_ok=True)

    # Run ffmpeg at 5 fps for this video
    cmd = [
        "ffmpeg",
        "-y",
        "-i", video_path,
        "-vf", "fps=5",
        os.path.join(out_dir, "frame_%04d.jpg"),
    ]
    print("Running:", " ".join(cmd))
    subprocess.run(cmd, check=True)
```

```python
import cv2
import os
from ultralytics import YOLO

model = YOLO("runs/detect/train2/weights/best.pt")

frames_root = "frames2"
output_root = "detections"

os.makedirs(output_root, exist_ok=True)

for video_folder in sorted(os.listdir(frames_root)):
    video_frames_dir = os.path.join(frames_root, video_folder)

    if not os.path.isdir(video_frames_dir):
        continue
    out_dir = os.path.join(output_root, video_folder)
    os.makedirs(out_dir, exist_ok=True)

    saved_count = 0
    frame_files = sorted(f for f in os.listdir(video_frames_dir) if f.endswith(".jpg"))

    for frame_file in frame_files:
        frame_path = os.path.join(video_frames_dir, frame_file)
        frame = cv2.imread(frame_path)
        if frame is None:
            continue

        results = model.predict(frame, conf=0.25, verbose=False)

        if results[0].boxes is not None and len(results[0].boxes) > 0:
            drone_boxes = results[0].boxes[results[0].boxes.cls == 1]
            if len(drone_boxes) > 0:
                cv2.imwrite(os.path.join(out_dir, frame_file), frame)
                saved_count += 1
```

```python
import cv2
import os
import subprocess
import tempfile
import numpy as np
from pathlib import Path
from filterpy.kalman import KalmanFilter
from scipy.optimize import linear_sum_assignment
from ultralytics import YOLO

WEIGHTS      = "runs/detect/train2/weights/best.pt"
FRAMES_ROOT  = "frames2"
OUTPUT_DIR   = "tracked_output"
DRONE_CLASS  = 1
CONF_THRESH  = 0.25
MAX_COAST    = 5
IOU_THRESH   = 0.30
FPS          = 5

BBOX_COLOR   = (  0, 255,   0)
COAST_COLOR  = (100, 200, 100)
TRAJ_COLOR   = (  0, 255, 255)
TEXT_COLOR   = (255, 255, 255)

os.makedirs(OUTPUT_DIR, exist_ok=True)

def make_kalman(cx: float, cy: float) -> KalmanFilter:
    kf = KalmanFilter(dim_x=2, dim_z=2)
    kf.F = np.eye(2)
    kf.H = np.eye(2)
    kf.R = np.eye(2) * 100.0
    kf.Q = np.eye(2) * 10.0
    kf.P = np.eye(2) * 100.0
    kf.x = np.array([[cx], [cy]])
    return kf
class Track:
    _counter = 0

    def __init__(self, cx, cy, bbox):
        Track._counter += 1
        self.id         = Track._counter
        self.kf         = make_kalman(cx, cy)
        self.trajectory = [(int(cx), int(cy))]
        self.last_bbox  = bbox
        self.coast      = 0
        self.active     = True

    def predict(self):
        self.kf.predict()
        self.coast += 1

    def update(self, cx, cy, bbox):
        self.kf.update(np.array([[cx], [cy]]))
        self.coast     = 0
        self.last_bbox = bbox
        self.trajectory.append(self._center())

    def record_coast(self):
        self.trajectory.append(self._center())

    def _center(self):
        return (int(self.kf.x[0, 0]), int(self.kf.x[1, 0]))

def box_iou(b1, b2):
    ix1, iy1 = max(b1[0], b2[0]), max(b1[1], b2[1])
    ix2, iy2 = min(b1[2], b2[2]), min(b1[3], b2[3])
    inter = max(0.0, ix2 - ix1) * max(0.0, iy2 - iy1)
    if inter == 0:
        return 0.0
    a1 = (b1[2]-b1[0]) * (b1[3]-b1[1])
    a2 = (b2[2]-b2[0]) * (b2[3]-b2[1])
    return inter / (a1 + a2 - inter)

def hungarian_match(tracks, detections, iou_thresh):
    if not tracks or not detections:
        return [], list(range(len(detections))), list(range(len(tracks)))
    cost = np.array(
        [[1.0 - box_iou(t.last_bbox, d[1]) for d in detections]
         for t in tracks]
    )
    ri, ci = linear_sum_assignment(cost)
    matched, used_d, used_t = [], set(), set()
    for r, c in zip(ri, ci):
        if cost[r, c] <= 1.0 - iou_thresh:
            matched.append((r, c))
            used_t.add(r); used_d.add(c)
    um_det = [i for i in range(len(detections)) if i not in used_d]
    um_trk = [i for i in range(len(tracks))     if i not in used_t]
    return matched, um_det, um_trk

def draw_bbox(frame, bbox, label, color, thickness=2):
    x1, y1, x2, y2 = (int(v) for v in bbox)
    cv2.rectangle(frame, (x1, y1), (x2, y2), color, thickness)
    cv2.putText(frame, label, (x1, max(y1-6, 14)),
                cv2.FONT_HERSHEY_SIMPLEX, 0.55, TEXT_COLOR, 1, cv2.LINE_AA)

def draw_trajectory(frame, trajectory):
    pts = [p for p in trajectory if p is not None]
    for i in range(1, len(pts)):
        cv2.line(frame, pts[i-1], pts[i], TRAJ_COLOR, 2, cv2.LINE_AA)
    if pts:
        cv2.circle(frame, pts[-1], 5, TRAJ_COLOR, -1)

def process_folder(folder_name: str, model: YOLO):
    frames_dir = os.path.join(FRAMES_ROOT, folder_name)
    frame_files = sorted(f for f in os.listdir(frames_dir) if f.endswith(".jpg"))

    if not frame_files:
        print(f"[SKIP] No frames found in {frames_dir}")
        return

    print(f"\n {folder_name}  |  {len(frame_files)} frames")

    tmp            = tempfile.mkdtemp(prefix=f"kf_{folder_name}_")
    tracks         = []
    saved          = []
    Track._counter = 0

    for idx, fname in enumerate(frame_files):
        frame = cv2.imread(os.path.join(frames_dir, fname))
        if frame is None:
            continue
        results = model.predict(frame, conf=CONF_THRESH, verbose=False)[0]
        dets = []
        if results.boxes is not None:
            for box in results.boxes:
                if int(box.cls[0]) != DRONE_CLASS:
                    continue
                x1, y1, x2, y2 = box.xyxy[0].tolist()
                dets.append((((x1+x2)/2, (y1+y2)/2), (x1, y1, x2, y2)))

        active = [t for t in tracks if t.active]
        for t in active:
            t.predict()
        matched, um_d, um_t = hungarian_match(active, dets, IOU_THRESH)

        for ti, di in matched:
            (cx, cy), bbox = dets[di]
            active[ti].update(cx, cy, bbox)
        for di in um_d:
            (cx, cy), bbox = dets[di]
            tracks.append(Track(cx, cy, bbox))
        for ti in um_t:
            t = active[ti]
            if t.coast <= MAX_COAST:
                t.record_coast()
            else:
                t.active = False
        visible = [t for t in tracks if t.active]
        if not visible:
            continue
        vis = frame.copy()
        for t in visible:
            if t.coast == 0:
                draw_bbox(vis, t.last_bbox, f"Drone #{t.id}", BBOX_COLOR)
            else:
                draw_bbox(vis, t.last_bbox,
                          f"Drone #{t.id} [coast {t.coast}/{MAX_COAST}]",
                          COAST_COLOR, thickness=1)
            draw_trajectory(vis, t.trajectory)

        cv2.putText(vis, f"Frame {idx}", (10, 24),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (200, 200, 200), 1)

        fpath = os.path.join(tmp, f"frame_{len(saved):06d}.jpg")
        cv2.imwrite(fpath, vis, [cv2.IMWRITE_JPEG_QUALITY, 95])
        saved.append(fpath)

    print(f"  Drone-present frames: {len(saved)} / {len(frame_files)}")

    if not saved:
        print("  [WARN] No drone detected — skipping output.")
        return
    out = os.path.join(OUTPUT_DIR, f"{folder_name}_tracked.mp4")
    cmd = [
        "ffmpeg", "-y",
        "-framerate", str(FPS),
        "-i", os.path.join(tmp, "frame_%06d.jpg"),
        "-c:v", "libx264", "-preset", "fast",
        "-crf", "20", "-pix_fmt", "yuv420p",
        out,
    ]
    r = subprocess.run(cmd, capture_output=True, text=True)
    if r.returncode != 0:
        print("  [ERROR] ffmpeg:\n", r.stderr[-600:])
    else:
        print(f"Saved → {out}  ({os.path.getsize(out)/1e6:.1f} MB)")

    for f in saved: os.remove(f)
    os.rmdir(tmp)

model = YOLO(WEIGHTS)

for folder in sorted(os.listdir(FRAMES_ROOT)):
    if os.path.isdir(os.path.join(FRAMES_ROOT, folder)):
        process_folder(folder, model)
```
